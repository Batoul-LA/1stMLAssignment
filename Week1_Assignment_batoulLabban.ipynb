```python
# 1. Load the dataset
import pandas as pd 
df = pd.read_csv("assignment_dataset/train.csv") 

# Display first 5 rows
df.head() 

# Display dataset shape
print(df.shape)

# Display column names
print(df.columns)

# Identify data types of each column
df.dtypes
```

```python
# 2. Missing Values Analysis

# Create a table listing:
# - Each column

# - Number of missing values

# - Percentage missing

miss_table = pd.DataFrame({
    " Column " : df.columns  ,
    " Missing Values" : df.isnull().sum(),
    " Percentage Missing " : (df.isnull().sum()/ len(df))*100
})  
print (miss_table)
```

# **⭐ Section 3 --- Data Cleaning (Code + Markdown)**

Perform and document:

```python
# 1. Handle Missing Values

# Use at least two different strategies, chosen appropriately:
# - mean/median imputation
# - mode imputation
# - indicator variable
# - row/column removal

df_cleand = df.copy()
df_cleand ['Age'] = df_cleand['Age'].fillna(df_cleand['Age'].mean())

df_cleand=df_cleand.dropna(subset=["Embarked"])

# Explain why each method is chosen.
# I used the technique "fill with mean value" on the feature 'Age' because it is a numerical variable
# I choosed the way of deleting rows of non_values of the column "Embarked" because it just has 2 non_values
```

```python
# 2. Verify that no missing values remain
df_cleand.info()
```

# **⭐ Section 4 --- Scaling (Code + Markdown)**

```python
# 1. Apply MinMax scaling to at least 2 numerical columns.

from sklearn.preprocessing import MinMaxScaler,StandardScaler
min_max=MinMaxScaler()
NUM_VALS=['Survived','Age']
df_MinMax =df.copy()
df_MinMax[NUM_VALS]=min_max.fit_transform(df_MinMax[NUM_VALS])

df_MinMax.head
```

```python
# 2. Apply Standardization to the same columns.

standard=MinMaxScaler()
NUM_VALS=['Survived','Age']
df_standard =df.copy()
df_standard[NUM_VALS]=standard.fit_transform(df_standard[NUM_VALS])

df_standard.head

```
